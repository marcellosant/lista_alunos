```python
import pandas as pd
```

```python
tipo_de_música = ['pop', 'rock', 'hip-hop', 'eletrônica', 'sertanejo']
```

```python
número_de_alunos = [35, 20, 15, 10, 20]
```

```python
dados = {'tipo_de_música':tipo_de_música, 'número_de_alunos':número_de_alunos}
```

```python
dados = pd.DataFrame(dados)
dados
```

```python
dados.columns
```

```python
dados = dados.loc[dados.index.repeat(dados.número_de_alunos)].reset_index(drop=True)
dados.drop(columns=['número_de_alunos'], inplace=True)
dados
```

```python
# frequência absoluta
dados.tipo_de_música.value_counts()
```

```python
# frequência relativa
dados.tipo_de_música.value_counts(normalize=True)
```

```python
# frequência relativa percentual
relativa = dados.tipo_de_música.value_counts(normalize=True) * 100
relativa
```

```python
# frequência relativa percentual
relativa = relativa.apply(lambda valor: f'{valor}%')
relativa
```

```python
# frequência cumulativa
dados.tipo_de_música.value_counts().cumsum()
```

```python
# número total de alunos
f'o número total de alunos é {dados.shape[0]}'
```

```python
dados['número_alunos'] = 1
dados
```

```python
dados_agrupados = dados.groupby(dados.tipo_de_música).sum()
dados_agrupados
```

```python
dados_agrupados.plot(kind='bar')
```

```python
dados_agrupados.plot(kind='line', marker='o')
```
